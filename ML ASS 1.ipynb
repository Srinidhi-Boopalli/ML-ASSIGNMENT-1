```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("WineQT.csv")
data
```

```python
data.head()
```

```python
data['quality'] = data['quality'].apply(lambda x: 0 if x <= 6 else 1)
```

```python
data['quality'].value_counts()
```

```python
 
features= data.drop('quality', axis= 1)
quality = data['quality']
data.head()
```

```python
scaler= MinMaxScaler()

data_scaled = scaler.fit_transform(features)
data_f=pd.DataFrame(data_scaled, columns=features.columns)
data_s= pd.concat([data_f, quality], axis=1)
```

```python
data_s.head()
```

```python
data_s.to_csv('WineQT_scaled.csv', index= False)
```
